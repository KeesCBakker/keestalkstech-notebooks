```python
import io
from urllib.request import urlopen, Request
from PIL import Image

def url_to_pil(url: str) -> PilImage:
    headers = { 'User-Agent': 'Mozilla/6.0'}
    data = urlopen(Request(url, headers=headers)).read()
    bts = io.BytesIO(data)
    return Image.open(bts)
```

```python
import matplotlib.pyplot as plt

def display_images(images, columns=4, figsize_multiplier=5):
    size=(len(images)/columns * figsize_multiplier, len(images)/columns * figsize_multiplier)
    plt.figure(figsize=size)
    for i, image in enumerate(images):
        plt.subplot(len(images) / columns + 1, columns, i + 1)
        plt.imshow(image)
```
